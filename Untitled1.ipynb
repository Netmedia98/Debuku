```python
# 1. Standard Library ("Bagian kode ini berfungsi untuk melakukan EDA (Exploratory Data Analysis). 
# Kita menyederhanakan data gambar yang kompleks agar bisa divisualisasikan secara sederhana 
# sehingga kita tahu apakah dataset kita sudah cukup bagus dan terorganisir sebelum melatih model AI.")
import os
import math
import copy
import random

# 2. Data Manipulation & Visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# 3. Utility & Progress
from tqdm.auto import tqdm

# 4. Machine Learning (Scikit-Learn)
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# 5. Deep Learning (PyTorch Core)
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from torch.cuda.amp import GradScaler, autocast

# 6. Computer Vision (Torchvision & Timm)
from torchvision import datasets, transforms
import timm
```

```python
# CONFIGURATION ("Bagian ini adalah konfigurasi dasar. Kita mengatur penggunaan GPU untuk kecepatan,
# menyeragamkan ukuran gambar, dan melakukan normalisasi agar model AI dapat belajar dengan lebih stabil dan efisien.")

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
DATA_DIR = '/content/drive/MyDrive/Fruit Freshness Dataset/Fruit Freshness Dataset'
SEED = 42
NUM_WORKERS = max(1, os.cpu_count() - 1)
IMG_SIZE = 224
BATCH_SIZE = 32
IMAGENET_DEFAULT_MEAN = [0.485, 0.456, 0.406]
IMAGENET_DEFAULT_STD = [0.229, 0.224, 0.225]
```

```python
#"Bagian ini berfungsi untuk mengunci acakan (reproducibility). Dengan menetapkan 'seed',
# saya memastikan bahwa hasil eksperimen ini konsisten dan dapat diulang kembali dengan hasil yang sama persis,
# baik di komputer saya maupun di komputer orang lain."
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)

print(f"Setup complete. Device set to: {DEVICE}")
```

```python
def find_classes_and_samples(root_dir):
    samples = []
    classes = set()

    if not os.path.exists(root_dir):
        print(f"Warning: Directory {root_dir} not found.")
        return [], []

    # Loop Level 1: Nama Buah (Apple, Banana, Strawberry)
    for fruit in sorted(os.listdir(root_dir)):
        fruit_path = os.path.join(root_dir, fruit)
        if os.path.isdir(fruit_path):
            # Loop Level 2: Kondisi (fresh, rotten)
            for condition in sorted(os.listdir(fruit_path)):
                cond_path = os.path.join(fruit_path, condition)
                if os.path.isdir(cond_path):
                    # Gabungkan nama buah dan kondisi menjadi nama kelas
                    class_name = f"{fruit}_{condition}"
                    classes.add(class_name)
                    

                    # Ambil semua gambar
                    for img_name in os.listdir(cond_path):
                        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                            img_path = os.path.join(cond_path, img_name)
                            samples.append((img_path, class_name))

    class_names = sorted(list(classes))
    class_to_idx = {cls_name: i for i, cls_name in enumerate(class_names)}
    final_samples = [(path, class_to_idx[name]) for path, name in samples]

    return final_samples, class_names
```

```python
#"Fungsi ini bertugas untuk scanning dataset. Ia menelusuri folder buah dan kondisinya,
# memberikan label otomatis untuk setiap gambar, dan mengubah label tersebut menjadi angka agar
# bisa diproses oleh algoritma pembelajaran mesin."
def train_transforms(image_size=IMG_SIZE):
    return transforms.Compose([
        transforms.RandomResizedCrop(image_size, scale=(0.8, 1.2)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.TrivialAugmentWide(),
        transforms.ToTensor(),
        transforms.Normalize(mean=IMAGENET_DEFAULT_MEAN, std=IMAGENET_DEFAULT_STD)
    ])

def val_transforms(image_size=IMG_SIZE):
    resize_dim = int(image_size * 1.14)
    return transforms.Compose([
        transforms.Resize(resize_dim, interpolation=transforms.InterpolationMode.BICUBIC),
        transforms.CenterCrop(image_size),
        transforms.ToTensor(),
        transforms.Normalize(mean=IMAGENET_DEFAULT_MEAN, std=IMAGENET_DEFAULT_STD)
    ])
```

```python
#"Class ini berfungsi sebagai jembatan yang memuat file gambar,menyeragamkan format warna ke RGB,
# dan menerapkan transformasi secara real-time agar data siap diumpankan ke dalam model untuk
# proses pembelajaran."
class Dataset(Dataset):
    def __init__(self, samples, transform=None):
        self.samples = samples
        self.transform = transform

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, target = self.samples[idx]
        try:
            image = Image.open(path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {path}: {e}")
            image = Image.new('RGB', (IMG_SIZE, IMG_SIZE))

        if self.transform:
            image = self.transform(image)

        return image, target
```

```python
#"Di bagian ini, saya membagi dataset menjadi dua: 80% untuk melatih model dan 20%
# untuk menguji performanya. Saya menggunakan teknik Stratified Split agar pembagian
# kategorinya tetap seimbang dan adil, sehingga model bisa belajar dari
# semua jenis buah secara merata."

all_samples, class_names = find_classes_and_samples(DATA_DIR)
all_targets = [s[1] for s in all_samples]

if len(all_samples) == 0:
    raise Exception("Error: No image found. recheck DATA_DIR path:", DATA_DIR)

# Stratified Split 80:20
train_samples, val_samples = train_test_split(
    all_samples,
    test_size=0.2,
    random_state=SEED,
    stratify=all_targets
)
```

```python
#"Baris ini menetapkan transformasi spesifik untuk masing-masing kumpulan data.
#Pada data Train, kita gunakan augmentasi agar model lebih tangguh, sedangkan
#pada data Validation, kita hanya melakukan standarisasi ukuran agar bisa diuji secara objektif."
train_dataset = Dataset(train_samples, transform=train_transforms(IMG_SIZE))
val_dataset = Dataset(val_samples, transform=val_transforms(IMG_SIZE))
```

```python
#"DataLoader berfungsi sebagai sistem pengiriman data yang efisien. Ia membagi gambar
# ke dalam beberapa kelompok kecil atau batch, mengacak urutan data latihan agar model
# tidak sekadar menghafal, dan menggunakan proses paralel agar proses training berjalan jauh lebih cepat."
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    persistent_workers=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    persistent_workers=True
)
```

```python
#"Baris ini berfungsi sebagai pemeriksaan terakhir untuk memvalidasi bahwa seluruh data telah terbaca dengan benar,
# jumlah pembagian datanya sudah tepat, dan sistem sudah mengenali seluruh kategori buah yang akan kita klasifikasikan."
print(f"Training Set : {len(train_dataset)} images")
print(f"Validation Set: {len(val_dataset)} images")
print(f"Total Classes : {len(class_names)}")
```

```python
#Bagian ini berperan untuk melakukan pemeriksaan distribusi data. Melalui grafik ini,
# kita dapat mengecek apakah kumpulan data tersebut seimbang atau tidak. Aspek ini sangat
# krusial untuk menjamin bahwa model AI yang kita buat nantinya akan adil dan tepat dalam
# mengidentifikasi setiap jenis buah tanpa berpihak pada satu kategori tertentu.
all_targets_indices = [s[1] for s in all_samples]

df_counts = pd.DataFrame({'label_idx': all_targets_indices})
df_counts['class_name'] = df_counts['label_idx'].map(lambda x: class_names[x])
class_counts = df_counts['class_name'].value_counts().sort_index()

# Plot Distribusi
plt.figure(figsize=(12, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='viridis')
plt.title('Jumlah Data per Kelas (Fruit Freshness)', fontsize=14)
plt.xlabel('Kategori Buah', fontsize=12)
plt.ylabel('Jumlah Gambar', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(f"Min samples: {class_counts.min()}")
print(f"Max samples: {class_counts.max()}")
```

```python
# Menampilkan Sample Gambar (Bagian ini dirancang untuk menampilkan koleksi contoh dari setiap kategori.
# Melalui tampilan ini, kita bisa secara langsung memastikan bahwa data yang diterima sudah tepat, labelnya akurat,
# dan mutu gambar yang akan dipelajari oleh model sudah memenuhi syarat.)
unique_samples = {}
# Cari satu contoh gambar untuk setiap kelas
for path, label_idx in all_samples:
    name = class_names[label_idx]
    if name not in unique_samples:
        unique_samples[name] = path 
    if len(unique_samples) == len(class_names):
        break

# Hitung grid plot
num_classes = len(class_names)
cols = 5
rows = math.ceil(num_classes / cols)

fig, axes = plt.subplots(rows, cols, figsize=(15, 4 * rows))
if num_classes > 1:
    axes = axes.flatten()
else:
    axes = [axes]

fig.suptitle('Sampel Gambar dari Setiap Kelas', fontsize=16)

sorted_samples = sorted(unique_samples.items())

for i in range(len(axes)):
    if i < len(sorted_samples):
        class_name, img_path = sorted_samples[i]
        try:
            img = Image.open(img_path)
            axes[i].imshow(img)
            axes[i].set_title(class_name, fontsize=10)
            axes[i].axis('off')
        except Exception as e:
            axes[i].text(0.5, 0.5, "Error Loading", ha='center')
    else:
        axes[i].axis('off') # Hide unused subplots

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Bagian ini merupakan jantung dari seluruh sistem. Fungsi train_model akan melaksanakan siklus
# pembelajaran di mana model berupaya meramalkan jenis buah, menilai kesalahannya, dan secara
# otomatis menyimpan versi terbaik dari dirinya sendiri berdasar pada akurasi tertinggi yang berhasil dicapai.
def get_data_setup(batch_size=32, test_size=0.2, image_size=224):
    all_samples, class_names = find_classes_and_samples(DATA_DIR)
    all_targets = [s[1] for s in all_samples]

    train_samples, val_samples = train_test_split(
        all_samples, test_size=test_size, random_state=SEED, stratify=all_targets
    )

    train_ds = Dataset(train_samples, transform=train_transforms(image_size))
    val_ds = Dataset(val_samples, transform=val_transforms(image_size))

    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True,
                              num_workers=NUM_WORKERS, persistent_workers=True, pin_memory=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False,
                            num_workers=NUM_WORKERS, persistent_workers=True, pin_memory=True)

    return train_loader, val_loader, class_names
def train_model(model, train_loader, val_loader, criterion, optimizer,
                scheduler=None, epochs=16, device=DEVICE, model_name="Model"):
    print(f"\n[Training] Starting: {model_name} | Epochs: {epochs}")
    save_filename = f"{model_name}.pth"
    model = model.to(device)
    scaler = torch.amp.GradScaler('cuda')
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    history = {'train_loss': [],
               'train_acc': [],
               'val_loss': [],
               'val_acc': []}

    for epoch in range(epochs):
        # Training
        train_loss = 0.0
        train_correct = 0
        train_total = 0

        model.train()
        pbar = tqdm(train_loader, desc=f"{model_name} Ep {epoch+1}/{epochs}", leave=False)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            with torch.amp.autocast('cuda'):
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            _, preds = torch.max(outputs, 1)
            train_loss += loss.item() * inputs.size(0)
            train_correct += torch.sum(preds == labels.data)
            train_total += labels.size(0)

            pbar.set_postfix(loss=loss.item())

        epoch_train_loss = train_loss / train_total
        epoch_train_acc = train_correct.double() / train_total

        # Validation
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                _, preds = torch.max(outputs, 1)
                val_loss += loss.item() * inputs.size(0)
                val_correct += torch.sum(preds == labels.data)
                val_total += labels.size(0)

        epoch_val_loss = val_loss / val_total
        epoch_val_acc = val_correct.double() / val_total

        # Get Current Learning Rate to display
        current_lr = optimizer.param_groups[0]['lr']

        # Scheduler Step
        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(epoch_val_loss)
            else:
                scheduler.step()

        # Save history
        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc.item())
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc.item())

        print(f"Train Loss: {epoch_train_loss:.4f} Train Acc: {epoch_train_acc:.4f} | "
              f"Val Loss: {epoch_val_loss:.4f} Val Acc: {epoch_val_acc:.4f} | "
              f"LR: {current_lr:.2e}")

        # Checkpoint
        if epoch_val_acc > best_acc:
            best_acc = epoch_val_acc
            torch.save(model.state_dict(), save_filename)
            print(f"Best {model_name} saved as {save_filename}")

    if os.path.exists(save_filename):
        print(f"Loading best weights from {save_filename}...")
        model.load_state_dict(torch.load(save_filename))
    else:
        print(f"[Warning] No model file saved. Check file {save_filename}")

    print(f"[Result] Final Best Val Acc: {best_acc:.4f}")
    return model, history
```

```python
#"Fungsi ini digunakan untuk mengevaluasi kesehatan model. Melalui grafik Loss dan Akurasi ini,
# kita bisa memantau apakah model berkembang dengan baik di setiap putarannya dan memastikan tidak
#  terjadi overfitting, yaitu kondisi di mana model hanya pintar di data latihan tapi gagal saat diuji dengan data baru."
def plot_results(history, title="Model Performance"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # Loss
    ax[0].plot(history['train_loss'], label='Train Loss')
    ax[0].plot(history['val_loss'], label='Val Loss')
    ax[0].set_title(f'{title} - Loss')
    ax[0].legend()
    ax[0].grid(True)

    # Accuracy
    ax[1].plot(history['train_acc'], label='Train Acc')
    ax[1].plot(history['val_acc'], label='Val Acc')
    ax[1].set_title(f'{title} - Accuracy')
    ax[1].legend()
    ax[1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
#Bagian ini merupakan evaluasi kinerja yang mendalam. Dengan memanfaatkan Laporan Klasifikasi,
# kita dapat mengamati rincian akurasi menurut masing-masing kelas. Sementara itu, melalui
# Matriks Kebingungan, kita dapat secara visual mencatat kategori mana yang sudah dikelompokkan
# dengan tepat dan kategori apa yang masih sering menyebabkan kebingungan atau kesalahan pada model.
def evaluate_model(model, loader, class_names, device=DEVICE):
    model.eval()
    y_true, y_pred = [], []

    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Evaluating"):
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    print("\nClassification Report:\n")
    print(classification_report(y_true, y_pred, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=False, cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title("Confusion Matrix")
    plt.xticks(rotation=90)
    plt.show()
```

```python
#Di tempat ini, saya menerapkan arsitektur Vision Transformer (ViT). Melalui pendekatan Transfer Learning,
# saya menggunakan model yang telah memiliki kecerdasan global untuk saya sesuaikan kembali (fine-tuning)
# sehingga bisa mengidentifikasi kesegaran buah dengan akurasi yang sangat tinggi.
BATCH_SIZE = 32
EPOCHS = 1
LR = 1e-5

train_loader, val_loader, class_names = get_data_setup(batch_size=BATCH_SIZE, test_size=0.2, image_size=224)
model = timm.create_model('vit_base_patch16_224', pretrained=True, num_classes=len(class_names))
```

```python
#Di langkah ini, saya memilih AdamW sebagai algoritma utama yang mengoptimalkan proses belajar model dan Cross
# Entropy sebagai ukuran untuk menilai kesalahannya. Dengan baris kode terakhir ini, kita secara resmi memulai
# tahapan pelatihan model Vision Transformer untuk mengenali tingkat kesegaran buah-buahan.
optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-2)
criterion = nn.CrossEntropyLoss()

trained_model, history = train_model(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        criterion=criterion,
        optimizer=optimizer,
        epochs=EPOCHS,
        device=DEVICE,
        model_name="vit"
    )
```
